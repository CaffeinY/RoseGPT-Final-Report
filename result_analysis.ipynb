```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def loadArray(name):
    arr = []
    with open(name, 'r') as file:
        for item in file:
            arr.append(float(item.split("\n")[0]))
    return arr
```

```python
train_loss_records_embed = np.load('final/train_loss_records_embedding.npy')
val_loss_records_embed = np.load('final/val_loss_records_embedding.npy')

train_loss_records_3FC_LeakyRelu = np.load('final/train_loss_records_3FC.npy')
val_loss_records_3FC_LeakyRelu = np.load('final/val_loss_records_3FC.npy')

train_loss_records_6FC = np.load('final/train_loss_records_6FC.npy')
val_loss_records_6FC = np.load('final/val_loss_records_6FC.npy')

train_loss_records_LR = np.load('final/train_loss_records_LR.npy')
val_loss_records_LR = np.load('final/val_loss_records_LR.npy')


```

```python
print(f"Final Train Loss of embed: {train_loss_records_embed[-1]}")
print(f"Best Validation Loss of embed: {min(val_loss_records_embed)}\n")


print(f"Final Train Loss of 6FC: {train_loss_records_6FC[-1]}")
print(f"Best Validation Loss of 6FC: {min(val_loss_records_6FC)}\n")

print(f"Final Train Loss of LR: {train_loss_records_LR[-1]}")
print(f"Best Validation Loss of LR: {min(val_loss_records_LR)}\n")
```
